### ARIMA

```python
import pmdarima as pm
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# Example: y = data["Prediction"]

# Step 1: hold out last n_periods for testing
n_periods = 10
train_y = y[:-n_periods]
test_y = y[-n_periods:]

# Step 2: Fit auto_arima on training data
model = pm.auto_arima(train_y,
                      seasonal=False,
                      stepwise=True,
                      trace=True)

print(model.summary())

# Step 3: Forecast next n_periods
forecast = model.predict(n_periods=n_periods)

# Step 4: Plot actual vs forecast
plt.figure(figsize=(10, 5))
plt.plot(y.index, y, label="Actual data")
plt.plot(test_y.index, forecast, color='red', label="Forecast on test")
plt.title("Auto ARIMA Forecast vs Actual")
plt.xlabel("Time")
plt.ylabel("Prediction")
plt.legend()
plt.show()

# Step 5: Print regression metrics
mae = mean_absolute_error(test_y, forecast)
mse = mean_squared_error(test_y, forecast)
rmse = np.sqrt(mse)
r2 = r2_score(test_y, forecast)

print("Regression metrics on test data:")
print(f"MAE: {mae:.2f}")
print(f"MSE: {mse:.2f}")
print(f"RMSE: {rmse:.2f}")
print(f"R² Score: {r2:.2f}")
```

### Prophet

```python
import pandas as pd
from prophet import Prophet
import matplotlib.pyplot as plt

# Example: your dataframe is `data`
# and it has columns: "Date" and "Prediction"

# Step 1: rename columns for Prophet
df = data[['Date', 'Prediction']].rename(columns={'Date': 'ds', 'Prediction': 'y'})

# Step 2: make sure 'ds' column is datetime
df['ds'] = pd.to_datetime(df['ds'])

# Step 3: initialize and fit Prophet model
model = Prophet()
model.fit(df)

# Step 4: make future dataframe (e.g., forecast next 10 days)
future = model.make_future_dataframe(periods=10)

# Step 5: forecast
forecast = model.predict(future)

# Step 6: plot forecast
fig1 = model.plot(forecast)
plt.title("Prophet Forecast")
plt.show()

# (Optional) plot components: trend, weekly/yearly seasonality
fig2 = model.plot_components(forecast)
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# Step 7: Merge actuals and predictions on the same dates
merged = pd.merge(df, forecast[['ds', 'yhat']], on='ds')

# Step 8: Calculate metrics
mae = mean_absolute_error(merged['y'], merged['yhat'])
mse = mean_squared_error(merged['y'], merged['yhat'])
rmse = np.sqrt(mse)
r2 = r2_score(merged['y'], merged['yhat'])

# Step 9: Print the metrics
print(f"📉 MAE:  {mae:.4f}")
print(f"📉 MSE:  {mse:.4f}")
print(f"📉 RMSE: {rmse:.4f}")
print(f"📈 R² Score: {r2:.4f}")
```

### LSTM MODEL

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split

# X: all your features (excluding Volume_AAPL)
# y: your target
X = data[['Close', 'High', 'Low', 'Open']]
y = data['Prediction']

# Scale features to [0,1] for LSTM (important!)
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# LSTM needs 3D input: (samples, timesteps, features)
# Let's use timesteps=1 for simplicity (can tune later)
X_scaled = np.reshape(X_scaled, (X_scaled.shape[0], 1, X_scaled.shape[1]))

# Split into train/test
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, shuffle=False)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

model = Sequential()
model.add(LSTM(units=50, return_sequences=False, input_shape=(X_train.shape[1], X_train.shape[2])))
model.add(Dense(1))  # output layer

model.compile(optimizer='adam', loss='mean_squared_error')
model.summary()
```

```python
history = model.fit(X_train, y_train, epochs=20, batch_size=32,
                    validation_data=(X_test, y_test))
```

```python
import matplotlib.pyplot as plt

# Predict
y_pred = model.predict(X_test)

# Plot actual vs predicted
plt.figure(figsize=(10,5))
plt.plot(y_test.values, label='Actual')
plt.plot(y_pred, label='Predicted')
plt.title('LSTM Prediction')
plt.xlabel('Samples')
plt.ylabel('Prediction')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
import numpy as np

# r2 score
r2 = r2_score(y_test, y_pred)

# mean squared error
mse = mean_squared_error(y_test, y_pred)

# mean absolute error
mae = mean_absolute_error(y_test, y_pred)

# root mean squared error
rmse = np.sqrt(mse)

# print nicely
print(f"R² Score: {r2:.4f}")
print(f"MSE: {mse:.4f}")
print(f"MAE: {mae:.4f}")
print(f"RMSE: {rmse:.4f}")
```

```python
import pandas as pd
import numpy as np
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import TimeSeriesSplit, RandomizedSearchCV
from xgboost import XGBRegressor
import matplotlib.pyplot as plt

# === STEP 1: Extract date features ===
def extract_date_features(df, col='Date'):
    df = df.copy()
    df['day'] = df[col].dt.day
    df['month'] = df[col].dt.month
    df['year'] = df[col].dt.year
    df['dayofweek'] = df[col].dt.dayofweek
    df['dayofyear'] = df[col].dt.dayofyear
    return df[['day', 'month', 'year', 'dayofweek', 'dayofyear']]

# === STEP 2: Load & prepare data ===
data['Date'] = pd.to_datetime(data['Date'])
X = extract_date_features(data, col='Date')
y = data['Prediction']

# === STEP 3: Scale data ===
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# === STEP 4: Train XGBoost with RandomizedSearchCV ===
param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.05, 0.1],
    'subsample': [0.7, 0.8, 1.0]
}

tscv = TimeSeriesSplit(n_splits=5)
xgb = XGBRegressor(objective='reg:squarederror', random_state=42)

search = RandomizedSearchCV(
    estimator=xgb,
    param_distributions=param_grid,
    n_iter=10,
    scoring='r2',
    n_jobs=-1,
    cv=tscv,
    verbose=0,
    random_state=42
)
# === Fit the model ===
search.fit(X_scaled, y)
print("✅ Model fitting done.")

# === Access the best estimator safely ===
best_model = search.best_estimator_

# === Evaluate model on training data ===
y_train_pred = best_model.predict(X_scaled)

r2 = r2_score(y, y_train_pred)
mse = mean_squared_error(y, y_train_pred)
mae = mean_absolute_error(y, y_train_pred)
rmse = np.sqrt(mse)

print(f"\n📊 Model Performance on Training Data:")
print(f"R² Score : {r2:.4f}")
print(f"MSE      : {mse:.4f}")
print(f"MAE      : {mae:.4f}")
print(f"RMSE     : {rmse:.4f}")

# === STEP 5: Take future dates as input ===
dates_input = input("Enter date(s) to predict, separated by commas (format: YYYY-MM-DD): ")
date_list = [d.strip() for d in dates_input.split(',')]
custom_dates = pd.to_datetime(date_list)

# === STEP 6: Predict for future dates ===
future_df = pd.DataFrame({'Date': custom_dates})
X_future = extract_date_features(future_df)
X_future_scaled = scaler.transform(X_future)
y_future_pred = search.best_estimator_.predict(X_future_scaled)

# === STEP 7: Show predictions ===
print("\n📈 XGBoost Predictions:")
for date, pred in zip(custom_dates, y_future_pred):
    print(f"Date: {date.date()} → Predicted: {pred:.2f}")

# === STEP 8: Plotting
plt.figure(figsize=(10, 5))
plt.plot(data['Date'], y, label='Historical', color='blue')
plt.scatter(custom_dates, y_future_pred, color='red', label='Future Predictions')
plt.title("XGBoost Predictions")
plt.xlabel("Date")
plt.ylabel("Predicted Value")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Hybrid model

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from tensorflow.keras.models import Model
from tensorflow.keras.layers import LSTM, Input
import tensorflow as tf

# === Prepare Data ===
X = data[['Close_AAPL', 'High_AAPL', 'Low_AAPL', 'Open_AAPL']]
y = data['Prediction']

# Scale for LSTM
lstm_scaler = MinMaxScaler()
X_scaled = lstm_scaler.fit_transform(X)

# Reshape for LSTM (samples, timesteps, features)
X_lstm = X_scaled.reshape(X_scaled.shape[0], 1, X_scaled.shape[1])

# Train/test split (no shuffle for time-series)
X_lstm_train, X_lstm_test, y_train, y_test = train_test_split(X_lstm, y, test_size=0.2, shuffle=False)

# === Define LSTM Encoder (not full model) ===
input_layer = Input(shape=(X_lstm_train.shape[1], X_lstm_train.shape[2]))
lstm_out = LSTM(units=50, return_sequences=False)(input_layer)
lstm_encoder = Model(inputs=input_layer, outputs=lstm_out)

# Extract temporal features
lstm_features_train = lstm_encoder.predict(X_lstm_train)
lstm_features_test = lstm_encoder.predict(X_lstm_test)

# === Train XGBoost on LSTM Features ===
xgb = XGBRegressor(
    subsample=0.8,
    n_estimators=300,
    min_child_weight=5,
    max_depth=8,
    learning_rate=0.05,
    gamma=0.2,
    colsample_bytree=1.0,
    objective='reg:squarederror',
    random_state=42
)

xgb.fit(lstm_features_train, y_train)

# === Prediction ===
y_pred = xgb.predict(lstm_features_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
r2=r2_score(y_test,y_pred)
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
print("R2:",r2)
print("MSE:", mse)
print("MAE:", mae)
```
